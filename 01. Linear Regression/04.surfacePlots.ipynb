## Surface Plots | Data Visualization

[Execute as py file to get interactive plots]

Surface Plots are used to
- Visualise Loss Functions in ML and DL.
- Visualise State or State Value Functions in RL

```python
import matplotlib.pyplot as plt
import numpy as np
```

a = [1,2,3] b = [4,5,6,7]
meshgrid(a,b) 
- will repeat the number of elements in a, b number of times along the rows.
- will repeat the number of elements in b, a number of times along the columns.

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])
a,b = np.meshgrid(a,b)
# b,a = np.meshgrid(b,a)
print(a)
print(b)
```

```python
#above we hae generated the data for meshgrid 
```

```python
fig = plt.figure()

#below axes object will give me 3d plot
axes = fig.gca(projection='3d')
##x,y,z
axes.plot_surface(a,b,a+b,cmap = 'rainbow')
plt.show()
```

```python
a = np.arange(-1,1,0.02)
b = a
a,b = np.meshgrid(a,b)
# b,a = np.meshgrid(b,a)
# print(a)
# print(b)
```

```python
fig = plt.figure()

#below axes object will give me 3d plot
axes = fig.gca(projection='3d')
##x,y,z
axes.plot_surface(a,b,a**2+b**2,cmap = 'rainbow')
plt.show()
```

## Contour Plots- 
### Cut the above plot with parallel planes and the regions with same value of z are denoted by same colors

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
plt.contour(a,b,a**2+b**2, cmap='rainbow')
plt.title('CONTOUR PLOT')
plt.show()
```
